```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_csv('training_data.csv')

# data.dropna(inplace=True)

# Display basic information about the dataset
print(data.info())

# Summary statistics
print(data.describe())
scaler = MinMaxScaler()
```

```python
# Convert 'timestamp' column to datetime format
data['timestamp'] = pd.to_datetime(data['timestamp'], utc=True)

# Extract month and day information from the timestamp
data['month'] = data['timestamp'].dt.month
data['day'] = data['timestamp'].dt.day
data['hour'] = data['timestamp'].dt.hour
```

```python
# Plot price against time in months
plt.figure(figsize=(12, 6))
plt.plot(data['timestamp'], data['price'], marker='.', linestyle='-', color='b')
plt.title('Electricity Spot Price Over Time (Monthly)')
plt.xlabel('Time')
plt.ylabel('Price ($/MWh)')
plt.grid(True)
plt.savefig('price_vs_time_monthly.png')  # Save the plot as an image
plt.show()
```

```python
# Plotting price against temperature on a monthly basis
plt.figure(figsize=(10, 6))

# Aggregate data by month: calculate average price and temperature for each month
monthly_data = data.groupby('month').mean()

# Iterate over all months (from 1 to 12) to ensure all months are considered
for month in range(1, 13):
    if month in monthly_data.index:
        plt.plot(month, monthly_data.loc[month, 'price'], marker='o', linestyle='-', color='b', label='Price ($/MWh)')
        plt.plot(month, monthly_data.loc[month, 'temp_air'], marker='s', linestyle='-', color='r', label='Temperature (°C)')
        plt.text(month, monthly_data.loc[month, 'price'] + 1, f"{monthly_data.loc[month, 'price']:.1f}", ha='center', va='top', color='b')
        plt.text(month, monthly_data.loc[month, 'temp_air'] - 1, f"{monthly_data.loc[month, 'temp_air']:.1f}", ha='center', va='bottom', color='r')

# Customize plot settings
plt.title('Electricity Spot Price vs. Air Temperature (Monthly)')
plt.xlabel('Month')
plt.ylabel('Value')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
# plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plotting price against DNI and GHI on a monthly basis
plt.figure(figsize=(12, 6))

# Plot price against DNI
plt.plot(monthly_data.index, monthly_data['price'], marker='o', linestyle='-', color='b', label='Price ($/MWh)')
plt.plot(monthly_data.index, monthly_data['dni'], marker='s', linestyle='-', color='r', label='DNI (W/m^2)')

# Annotate mean values of price and DNI
for month in monthly_data.index:
    plt.text(month, monthly_data.loc[month, 'price'] + 1, f"{monthly_data.loc[month, 'price']:.1f}", ha='center', va='top', color='b')
    plt.text(month, monthly_data.loc[month, 'dni'] - 10, f"{monthly_data.loc[month, 'dni']:.1f}", ha='center', va='bottom', color='r')

plt.title('Electricity Spot Price vs. Direct Normal Irradiance (DNI) (Monthly)')
plt.xlabel('Month')
plt.ylabel('Value')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend()
plt.grid(True)
plt.show()
```

```python

# Plotting price against GHI
plt.figure(figsize=(12, 6))

plt.plot(monthly_data.index, monthly_data['price'], marker='o', linestyle='-', color='b', label='Price ($/MWh)')
plt.plot(monthly_data.index, monthly_data['ghi'], marker='s', linestyle='-', color='g', label='GHI (W/m^2)')

# Annotate mean values of price and GHI
for month in monthly_data.index:
    plt.text(month, monthly_data.loc[month, 'price'] + 1, f"{monthly_data.loc[month, 'price']:.2f}", ha='center', color='b')
    plt.text(month, monthly_data.loc[month, 'ghi'] - 10, f"{monthly_data.loc[month, 'ghi']:.2f}", ha='center', color='g')

plt.title('Electricity Spot Price vs. Global Horizontal Irradiance (GHI) (Monthly)')
plt.xlabel('Month')
plt.ylabel('Value')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plotting price against demand on a monthly basis
plt.figure(figsize=(10, 6))

# Plot price against demand
plt.plot(monthly_data.index, monthly_data['price'], marker='o', linestyle='-', color='b', label='Price ($/MWh)')
plt.plot(monthly_data.index, monthly_data['demand'], marker='s', linestyle='-', color='r', label='Demand (MW)')

# Annotate mean values of price and demand
for month in monthly_data.index:
    plt.text(month, monthly_data.loc[month, 'price'] + 1, f"{monthly_data.loc[month, 'price']:.2f}", ha='center', color='b')
    plt.text(month, monthly_data.loc[month, 'demand'] - 1, f"{monthly_data.loc[month, 'demand']:.2f}", ha='center', color='r')

plt.title('Electricity Spot Price vs. Demand (Monthly)')
plt.xlabel('Month')
plt.ylabel('Value')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plotting price against pv_power on a monthly basis
plt.figure(figsize=(10, 6))

# Plot price against pv_power
plt.plot(monthly_data.index, monthly_data['price'], marker='o', linestyle='-', color='b', label='Price ($/MWh)')
plt.plot(monthly_data.index, monthly_data['pv_power'], marker='s', linestyle='-', color='g', label='PV Power (kW)')

# Annotate mean values of price and pv_power
for month in monthly_data.index:
    plt.text(month, monthly_data.loc[month, 'price'] + 1, f"{monthly_data.loc[month, 'price']:.2f}", ha='center', color='b')
    plt.text(month, monthly_data.loc[month, 'pv_power'] - 1, f"{monthly_data.loc[month, 'pv_power']:.2f}", ha='center', color='g')

plt.title('Electricity Spot Price vs. PV Power (Monthly)')
plt.xlabel('Month')
plt.ylabel('Value')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend()
plt.grid(True)
plt.show()
```

```python
data['year'] = data['timestamp'].dt.year 
# Calculate monthly minimum temperature
monthly_min_temp = data.groupby(['year', 'month'])['temp_min'].min().reset_index()

# Aggregate data by month: calculate average price for each month
monthly_avg_price = data.groupby(['year', 'month'])['price'].mean().reset_index()

# Merge monthly minimum temperature and average price data
monthly_data = pd.merge(monthly_avg_price, monthly_min_temp, on=['year', 'month'])

# Plotting price against minimum temperature of each month
plt.figure(figsize=(10, 6))

# Plot price against minimum temperature
plt.plot(monthly_data['month'], monthly_data['price'], marker='o', linestyle='-', color='b', label='Price ($/MWh)')
plt.plot(monthly_data['month'], monthly_data['temp_min'], marker='s', linestyle='-', color='r', label='Min Temperature (°C)')

# Annotate mean values of price and minimum temperature
for index, row in monthly_data.iterrows():
    plt.text(row['month'], row['price'] + 1, f"{row['price']:.2f}", ha='center', color='b')
    plt.text(row['month'], row['temp_min'] - 1, f"{row['temp_min']:.2f}", ha='center', color='r')

plt.title('Electricity Spot Price vs. Minimum Air Temperature (Monthly)')
plt.xlabel('Month')
plt.ylabel('Value')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculate monthly maximum temperature
monthly_max_temp = data.groupby(['year', 'month'])['temp_max'].max().reset_index()

# Aggregate data by month: calculate average price for each month
monthly_avg_price = data.groupby(['year', 'month'])['price'].mean().reset_index()

# Merge monthly maximum temperature and average price data
monthly_data = pd.merge(monthly_avg_price, monthly_max_temp, on=['year', 'month'])

# Plotting price against maximum temperature of each month
plt.figure(figsize=(10, 6))

# Plot price against maximum temperature
plt.plot(monthly_data['month'], monthly_data['price'], marker='o', linestyle='-', color='b', label='Price ($/MWh)')
plt.plot(monthly_data['month'], monthly_data['temp_max'], marker='s', linestyle='-', color='r', label='Max Temperature (°C)')

# Annotate mean values of price and maximum temperature
for index, row in monthly_data.iterrows():
    plt.text(row['month'], row['price'] + 1, f"{row['price']:.2f}", ha='center', color='b')
    plt.text(row['month'], row['temp_max'] - 1, f"{row['temp_max']:.2f}", ha='center', color='r')

plt.title('Electricity Spot Price vs. Maximum Air Temperature (Monthly)')
plt.xlabel('Month')
plt.ylabel('Value')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Filter data for May (month = 5)
may_data = data[(data['month'] == 5)]

# Aggregate temperature data by day in May
daily_avg_temp = may_data.groupby('day')['temp_air'].mean()

# Aggregate price data by day in May
daily_avg_price = may_data.groupby('day')['price'].mean()

# Plotting price against average daily temperature for May
plt.figure(figsize=(10, 6))

# Scatter plot of price against average daily temperature
plt.scatter(daily_avg_temp, daily_avg_price, color='b', alpha=0.5)

plt.title('Electricity Spot Price vs. Average Daily Temperature in May')
plt.xlabel('Average Daily Temperature (°C)')
plt.ylabel('Price ($/MWh)')
plt.grid(True)
plt.show()
```

```python
# Filter data for May (month = 5)
month_data = data[data['month'] == 5]

# Aggregate price, pv_power, and temperature data by hour of the day in May
hourly_avg_price = month_data.groupby('hour')['price'].mean()
hourly_avg_pv_power = month_data.groupby('hour')['pv_power'].mean()
hourly_avg_temp = month_data.groupby('hour')['temp_air'].mean()

# Normalize the values for annotations
price_min = hourly_avg_price.min()
price_max = hourly_avg_price.max()
pv_power_min = hourly_avg_pv_power.min()
pv_power_max = hourly_avg_pv_power.max()

hourly_avg_price_norm = scaler.fit_transform(hourly_avg_price.values.reshape(-1, 1))
hourly_avg_pv_power_norm = scaler.fit_transform(hourly_avg_pv_power.values.reshape(-1, 1))
hourly_avg_temp_norm = scaler.fit_transform(hourly_avg_temp.values.reshape(-1, 1))

# Plotting price, pv_power, and temperature on each hour of the day in May
plt.figure(figsize=(12, 6))

# Line plot of price, pv_power, and temperature (using normalized values)
plt.plot(range(24), hourly_avg_price_norm, marker='o', linestyle='-', color='b', label='Price ($/MWh)')
plt.plot(range(24), hourly_avg_pv_power_norm, marker='s', linestyle='-', color='g', label='PV Power (kW)')
plt.plot(range(24), hourly_avg_temp_norm, marker='^', linestyle='-', color='r', label='Temperature (normalized)')

# Annotate price values
for hour, price_norm in zip(range(24), hourly_avg_price_norm):
    plt.text(hour, price_norm, f"{price_norm[0]:.2f}", ha='center', va='bottom', color='b')

# Annotate pv_power values
for hour, pv_power_norm in zip(range(24), hourly_avg_pv_power_norm):
    plt.text(hour, pv_power_norm, f"{pv_power_norm[0]:.2f}", ha='center', va='bottom', color='g')

# Annotate temperature values
for hour, temp_norm in zip(range(24), hourly_avg_temp_norm):
    plt.text(hour, temp_norm, f"{temp_norm[0]:.2f}", ha='center', va='bottom', color='r')

plt.title('Normalized Electricity Spot Price, PV Power, and Temperature vs. Hour of the Day in May')
plt.xlabel('Hour of the Day')
plt.ylabel('Normalized Value')
plt.xticks(range(24))
plt.legend()
plt.grid(True)
plt.show()
```
